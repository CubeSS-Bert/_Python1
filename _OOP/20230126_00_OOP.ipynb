```python
class ChessSet:
    
    def __init__(self):
        # self.set_board()
        self.board = Board()
        self.pieces = []
        self.populate_list()
        
    # def set_board(self):
        # self.board = Board()
    
    def populate_list(self):
        colors = ["White", "Black"]
        for color in colors:
            for x in range(32):
                if x % 2 == 0:
                    self.pieces.append(King(color, self))
                else:
                    self.pieces.append(Rook(color, self))
                
class Board:
    
    def __init__(self) -> None:
        pass
    

class Piece:
    
    def __init__(self, color, chess_set):
        self.color = color
        self.chess_set = chess_set
    
    def move(self):
        pass


# Inheritance > King van Piece
class King(Piece):
    
    def __init__(self, color, chess_set):
        super().__init__(color, chess_set)
        self.shape = "K"    
    
    def move(self):
        print('Alle kanten, 1 veld')

    
# Inheritance > Rook van Piece
class Rook(Piece):
    
    def __init__(self, color, chess_set):
        super().__init__(color, chess_set)
        self.shape = "R"    
    
    def move(self):
        print('Rechtdoor, 1 veld')
    
    
```
