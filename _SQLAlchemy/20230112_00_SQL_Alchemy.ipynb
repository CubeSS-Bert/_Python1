```python
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:Syntra_1408@localhost: 5432/dvdrental')
```

```python
from sqlalchemy import MetaData
metadata = MetaData()
```

```python
from sqlalchemy import Table, Column, Integer, Numeric, String, ForeignKey

cookies = Table('cookies', metadata,
    Column('cookie_id', Integer(), primary_key=True),
    Column('cookie_name', String(50), index=True),
    Column('cookie_recipe_url', String(255)),
    Column('cookie_sku', String(55)),
    Column('quantity', Integer()),
    Column('unit_cost', Numeric(12, 2))
)
```

```python
from datetime import datetime
from sqlalchemy import DateTime

users = Table('users', metadata,
    Column('user_id', Integer(), primary_key=True),
    Column('username', String(15), nullable=False, unique=True),
    Column('email_address', String(255), nullable=False),
    Column('phone', String(20), nullable=False),
    Column('password', String(25), nullable=False),
    Column('created_on', DateTime(), default=datetime.now),
    Column('updated_on', DateTime(), default=datetime.now, onupdate=datetime.now),
    )
```

```python
from sqlalchemy import PrimaryKeyConstraint, UniqueConstraint, CheckConstraint

PrimaryKeyConstraint('user_id', name='user_pk')
UniqueConstraint('username', name='uix_username')
CheckConstraint('unit_cost >= 0.00', name='unit_cost_positive')
```

```python
from sqlalchemy import ForeignKey, Boolean

orders = Table('orders', metadata,
    Column('order_id', Integer(), primary_key=True),
    Column('user_id', ForeignKey('users.user_id')),
    Column('shipped', Boolean(), default=False)
)

line_items = Table('line_items', metadata,
    Column('line_items_id', Integer(), primary_key=True),
    Column('order_id', ForeignKey('orders.order_id')),
    Column('cookie_id', ForeignKey('cookies.cookie_id')),
    Column('quantity', Integer()),
    Column('extended_cost', Numeric(12, 2))
)
```

```python
def get_orders_by_customers(cust_name):
    colums = [orders.c.order_id, users.c.username, cookies.c.cookie_name, line_items.c.quantity, line_items.c.extended_cost]
    cust_orders = select(colums)
    cust_orders = cust_orders.select_from(users.join(orders).join(line_items).join(cookies))
    cust_orders.where(users.c.username == cust_name)
    result = engine.execute(cust_orders).fetchall()
    return result

get_orders_by_customers('cakeeater')


    
```

```python
def get_shipped_orders_by_customers(cust_name, shipped=None, details=False):
    colums = [orders.c.order_id, users.c.username, cookies.c.cookie_name, users.c.phone]
    joins = users.join(orders)
    
    if details:
        colums.extend([cookies.c.cookie_name, line_items.c.quantity, line_items.c.extended_cost])
        joins = joins.join(line_items).join(cookies)
        
        
    cust_orders = select(colums)
    cust_orders = cust_orders.select_from(joins)
    cust_orders.where(users.c.username == cust_name)
    
    if shipped is not None:
        cust_orders.where(users.c.username == cust_name)
        
    
    
    result = engine.execute(cust_orders).fetchall()
    return result

get_shipped_orders_by_customers('cakeeater')

```
